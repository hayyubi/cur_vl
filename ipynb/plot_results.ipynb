```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)

x = [0.1, 0.25, 0.5, 1.0]
xi = list(range(len(x)))

y_bbox = [53.73, 51.74, 50.29, 49.44]
y_bbox_cur = [53.70, 53.74, 52.44, 47.81]

plt.plot(xi, y_bbox, label='BBox', marker='o')
plt.plot(xi, y_bbox_cur, label='BBox+Cur', marker='o')
plt.xticks(xi, x)
for i, v in enumerate(y_bbox):
    ax.text(i, v-0.5, "%.2f" %v, ha="center")
for i, v in enumerate(y_bbox_cur):
    ax.text(i, v+0.1, "%.2f" %v, ha="center")   
    
plt.xlabel('Datasize Percent')
plt.ylabel('Class wise accuracy')
plt.title('COCO GT: Accuracy Vs Datasize')
plt.legend()
plt.show()
```

```python
def plot(x, y_bbox, y_bbox_cur, x_label, y_label, title):
    fig = plt.figure(figsize=(5,2))
    ax = fig.add_subplot(111)
    
    xi = list(range(len(x)))
    
    plt.plot(xi, y_bbox, label='OVR-CNN$_O$', marker='o')
    plt.plot(xi, y_bbox_cur, label='Ours', marker='o')
    plt.xticks(xi, x)
    
    ax.text(0+0.1, y_bbox[0]-2.2, "%.2f" %y_bbox[0], ha="center")
    for i, v in enumerate(y_bbox):
        if i==0 or i==len(y_bbox)-1:
            continue
        ax.text(i, v-2.2, "%.2f" %v, ha="center")
    ax.text(i-0.4, y_bbox[-1]-0.4, "%.2f" %y_bbox[-1], ha="center")
    
    ax.text(0+0.4, y_bbox_cur[0]-0.6, "%.2f" %y_bbox_cur[0], ha="center")
    for i, v in enumerate(y_bbox_cur):
        if i==0:
            continue
        ax.text(i, v+1.2, "%.2f" %v, ha="center")   
        
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()
```

```python
def plot_bars(values, x_ticklabel, x_label, y_label, title, font_scale=1, rot_angle=0):

    x = np.arange(len(x_ticklabel)) - 0.5 * np.arange(len(x_ticklabel)) # the label locations
    width = 0.35  # the width of the bars

    with sns.plotting_context("notebook", font_scale=font_scale):
        fig, ax = plt.subplots(figsize=(4,3))
        ax.grid(zorder=0)
        rects1 = ax.bar(x , values, width, label=None, zorder=3)
    
        # Add some text for labels, title and custom x-axis tick labels, etc.
        ax.set_ylabel(y_label)
        ax.set_xlabel(x_label)
        ax.set_title(title)
        ax.set_xticks(x)
        ax.set_xticklabels(x_ticklabel)
        ax.legend()
    
    
        def autolabel(rects):
            """Attach a text label above each bar in *rects*, displaying its height."""
            for i, rect in enumerate(rects):
                if i==len(rects)-1:
                    continue
                height = rect.get_height()
                ax.annotate('{}'.format(height),
                            xy=(rect.get_x() + rect.get_width() / 2, height),
                            xytext=(0, 3),  # 3 points vertical offset
                            textcoords="offset points",
                            ha='center', va='bottom')
    
    
        autolabel(rects1)
        for item in ax.get_xticklabels():
            item.set_rotation(rot_angle)
    
        fig.tight_layout()
    
        plt.show()
```

```python
values = [3158, 64382, 221819, 302380]
x_ticklabel = ['Phase' + str(i) for i in list(range(1,5))]

plot_bars(values, x_ticklabel, '', 'Number of Captions',
          '')
```

```python
def plot_bars(values, x_ticklabel, x_label, y_label, title, font_scale=1, rot_angle=0):

    x = np.arange(len(x_ticklabel)) - 0.5 * np.arange(len(x_ticklabel)) # the label locations
    width = 0.35  # the width of the bars

    with sns.plotting_context("notebook", font_scale=font_scale):
        fig, ax = plt.subplots(figsize=(4,3))
        ax.grid(zorder=0)
        rects1 = ax.bar(x , values, width, label=None, zorder=3, color='green')
    
        # Add some text for labels, title and custom x-axis tick labels, etc.
        ax.set_ylabel(y_label)
        ax.set_xlabel(x_label)
        ax.set_title(title)
        ax.set_xticks(x)
        ax.set_xticklabels(x_ticklabel)
        ax.legend()
    
    
        def autolabel(rects):
            """Attach a text label above each bar in *rects*, displaying its height."""
            for i, rect in enumerate(rects):
                if i==0:
                    continue
                height = rect.get_height()
                ax.annotate('{}'.format(height),
                            xy=(rect.get_x() + rect.get_width() / 2, height),
                            xytext=(0, 3),  # 3 points vertical offset
                            textcoords="offset points",
                            ha='center', va='bottom')
    
    
        autolabel(rects1)
        for item in ax.get_xticklabels():
            item.set_rotation(rot_angle)
    
        fig.tight_layout()
    
        plt.show()
```
